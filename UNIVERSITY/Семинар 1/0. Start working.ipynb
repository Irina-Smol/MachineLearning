#### Основные команды Jupyter notebook:
- A - создать ячейку выше
- B - создать ячейку ниже
- DD - удалить ячейку
- Ctrl + Enter - выполнить ячейку и остаться в ней
- Shift + Enter - выполнить ячейку и перейти в следующую, если нет следующей то создаёт
- Ctrl + . - закомментировать / раскомментировать выделенный фрагмент
- M - сменить тип ячейки на markdown
- Y - сменить тип ячейки на код
- Ctrl+Shift+M - объединить выделенные ячейки
- Ctrl+Shift+- - разделить ячейку
- Ctrl_Shift+H - справка по клавишам

#### Синтаксис markdown-ячеек:
- символ "#" - уровень заголовка
- символ "-" или "*" - список элементов, с отступом Tab - уровни списка
- символы 1. , 2. и т.д. - нумерованный список
- символы "**" в начале и в конце - жирный текст
- символ "*" в начале и в конце - курсив
- символы "~~" в начале и в конце - зачёркнутый
- символы "---" - горизонтальная линия
