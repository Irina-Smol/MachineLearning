```python
import numpy as np 
import pandas as pd
```

```python
breast_cancer_df = pd.read_csv("data (1).csv")
breast_cancer_df
```

```python
breast_cancer_df.info()
```

```python
breast_cancer_df.isnull().sum() # проверка на пропущенные значения
```

```python
breast_cancer_df.drop(['Unnamed: 32', 'id'], axis = 1, inplace = True) # удалить ненужные столбцы
```

```python
from sklearn.preprocessing import LabelEncoder # преобразовать категориальные данные в числа
le = LabelEncoder()
breast_cancer_df['diagnosis'] = le.fit_transform(breast_cancer_df['diagnosis'])
```

```python
breast_cancer_df
```

```python
# обозначаем зависимую и независимую переменные
y = breast_cancer_df.diagnosis.values
x = breast_cancer_df.drop(['diagnosis'], axis = 1)
```

```python
# разделяем датасет на данные для обучения и тренировки модели
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.15, random_state = 42)
```

```python
# для стандартизации данных
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# обучаем модель
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski',p = 2)
knn.fit(x_train,y_train)

# строим confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print('confusion matrix:\n',cm)

# проверяем accuracy
from sklearn.metrics import accuracy_score
knna = accuracy_score(y_test,y_pred)
print('accuracy score = ',accuracy_score(y_test,y_pred))
```
