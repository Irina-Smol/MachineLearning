# Training the model

```python
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'radius_se', 'area_se', 'radius_worst', 'perimeter_worst', 
                   'compactness_worst']
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.15, random_state=1)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis']
test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_depth': (1, 2, 3, 4), 'n_estimators': (10, 50, 100, 500)}
best_model = GridSearchCV(model, parameters)
```

```python
best_model.fit(train_x, train_y)
```

```python
best_model.best_params_
```

```python
model.fit(train_x, train_y)
```

```python
predictions = best_model.predict(test_x)
```

```python
test_y
```

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(test_y, predictions)
```

# Other metrics: precision, recall, accuracy

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(test_y, predictions)
print('The precision is %.2f' % precision)

recall = recall_score(test_y, predictions)
print('The precision is %.2f' % recall)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test_y, predictions)
print('The precision is %.2f' % accuracy)
```
